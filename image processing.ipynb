```python
#####MORPHOLOGICAL ANALYSIS
#--erosion --loss of pixels from boundary
#--dilation --gain of pixels to boundary
```

```python
img=cv2.imread(r"C:\Users\USER\AppData\Local\Temp\image_data\np2.jpg")
plt.imshow(img,cmap='gray')
plt.show()
```

```python
kernel=np.ones((7,7))#from boundaries the pixel loss is there which are very thin objects
img2=cv2.erode(img,kernel)
plt.imshow(img2,cmap='gray')
plt.show()
```

```python
img3=cv2.dilate(img2,kernel)#the image is regained from the same kernel which are eroded from the image
plt.imshow(img3,cmap='gray')
plt.show()
```

```python
def color_detect(img):
    img2=img.copy()
    m,n,q=img.shape
    for i in range(m):
        for j in range(n):
            if img[i,j,1]<100 and img[i,j,1]<100 and img[i,j,2]<100:
                img2[i,j,0]=255
                img2[i,j,1]=255
                img2[i,j,2]=255 
    return img2
```

```python
#face detection
fd=cv2.CascadeClassifier(r"C:\Users\USER\AppData\Local\Temp\image_data\haarcascade_frontalface_alt.xml")
img=im.imread(r"C:\Users\USER\AppData\Local\Temp\image_data\people.jpg")
plt.imshow(img)#1st line is the trained model of classification
plt.show()
```

```python
corners=fd.detectMultiScale(img)#detect images of multiple faces of th
corners
```

```python
for x,y,w,h in corners:
    cv2.rectangle(img,(x,y),(x+w,y+h),[0,0,255],8)
plt.imshow(img)#draws the corners of multiple rectangles of faces 
plt.show()
```

```python
#working with video detecting faces like the image
vid=cv2.VideoCapture(0)
while True:
    flag,img=vid.read()#taking the access of the webcam
    corners=fd.detectMultiScale(img)
    for x,y,w,h in corners:
        cv2.rectangle(img,(x,y),(x+w,y+h),[0,0,255],2)
    cv2.imshow("imshow",img)
    if cv2.waitKey(1) & 0xFF==ord('q'):#waitkey checks in w=every millisec q is pressed or not
        break#q is pressed then the webcam is released from python 
vid.release()
cv2.destroyAllWindows()
```

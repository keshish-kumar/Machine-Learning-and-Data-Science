```python
#fetching the data from the url of the machines which are used in the factory
import pandas as pd
df=pd.read_csv(r"https://raw.githubusercontent.com/anshupandey/Machine_Learning_Training/master/datasets/maintenance_data.csv")
```

```python
#finding the unique values per provider column
df.provider.unique()
```

```python
#finding the unique values per team column
df.team.unique()
```

```python
#data cleaning started by finding the duplicate values if existed
df.duplicated().sum()
```

```python
#finding the null values if existed
df.isnull().sum()#we got the null values in temperatureInd & pressureInd and then:
```

```python
df['pressureInd'].fillna(df['pressureInd'].mean(),inplace=True)
df.fillna(df.median(),inplace=True)#removing the null values through mean and median
```

```python
df.isnull().sum()#after finding the null clearing it as null
```

```python
df.duplicated().sum()#duplicated values is also checked as zero
```

```python
#moisture solving the problem of 75% and max value is error is coming
df.moistureInd.quantile(0.999)
```

```python
df.shape#before quantile used
```

```python
df=df[df.moistureInd<=df.moistureInd.quantile(0.999)]
df.shape#after quantile used
```

```python
df.describe()#
```

```python
#moisture prblem is solved
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,5))
sns.distplot(df.moistureInd[df.broken==0])
sns.distplot(df.moistureInd[df.broken==1])
plt.legend(['0','1'])
plt.show()#the below plot is shown as that moisture has a less effect on the lifestime of the machines
#as the moisture increases the machines gets broke.
```

```python
plt.figure(figsize=(12,5))
sns.countplot(df.provider,order=df.provider.unique())
plt.show()#this plot shows about the approximation from various providers who had supplied the machine
plt.figure(figsize=(12,5))
sns.countplot(df.provider[df.broken==1],order=df.provider.unique())
plt.show()#this plot shows the broken machines record approximation from various providers
#when we compare the various comparison from overall machines and broken machines where provider 3 gives the worst service when checked
```

```python
plt.figure(figsize=(12,5))
sns.countplot(df.team,order=df.team.unique())
plt.show()#total optimisation done by the team of all machines supplied
plt.figure(figsize=(12,5))
sns.countplot(df.team[df.broken==1],order=df.team.unique())
plt.show()#compared with the 1st plot then broken status is recorded
```

```python
plt.figure(figsize=(12,5))
sns.pointplot(x='provider',y='lifetime',hue='broken',data=df)
plt.show()#after getting the plot for provider 3 service to be worst then in this we can find out the machines:
#that 3's machines are down by the in between 60-70 as compared to other provider 
```

```python
plt.figure(figsize=(12,5))
sns.pointplot(x='team',y='lifetime',hue='broken',data=df)
plt.show()#optimasation can be shown that the machines from 3 are the worst having less lifetime
```

```python
#categorical plot
#categorical vs categorical
plt.figure(figsize=(8,4))
sns.catplot(x='provider',col='team',row='broken',kind='count',data=df)
plt.show()#this shows the complete overview of all providers in categorical plot that 3's sevice is less than the other twos
```
